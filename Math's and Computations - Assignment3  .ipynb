```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def flipping_a_coin(Num_Of_Trials):
    values = ["Head","Tail"]
    probs = [0.7,0.3]

# Draw a weighted sample
    sample = np.random.choice(values, Num_Of_Trials, p=probs)

# Numpy arrays containing counts for each result
    result, count = np.unique(sample, return_counts=True)
    probs = count / len(sample)
# Plot the results
    sns.barplot(x=result, y=probs)
    plt.title(
    "Discrete Probability Distribution for tossing a biased coin ({} Trials)\nNumber of heads: {} Number of tails: {}".format(Num_Of_Trials,count[0],count[1]))
    plt.ylabel("Probability")
    plt.xlabel("Outcome")
    plt.legend(probs)
    plt.show()
```

```python
flipping_a_coin(20)
```
